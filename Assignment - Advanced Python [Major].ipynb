### Topic: To visualize how honey production has changed over the years (1998-2021) in the United States.
<b>Background:</b><br>
In 2006, global concern was raised over the rapid decline in the honeybee population, an integral component of American honey agriculture. Large numbers of hives were lost to Colony Collapse Disorder, a phenomenon of disappearing worker bees causing the
remaining hive colony to collapse. Speculation as to the cause of this disorder points to hive diseases and pesticides harming the pollinators, though no overall consensus has been reached. The U.S. used to locally produce over half the honey it consumes per year. Now, honey mostly comes from overseas, with 350 of the 400 million pounds of honey consumed every year originating from imports.This dataset provides insight into honey production supply and demand in America from 1998 to 2021.

#### <b>Objective:</b><br>
To visualize how honey production has changed over the years
(1998-2021) in the United States.

### <b>Key questions to be answered:</b><br>
<b>Q1)</b> How has honey production yield changed from 1998 to 2021?<br>
<b>Q2)</b> Over time, what are the major production trends across the states?<br>
<b>Q3)</b> Does the data show any trends in terms of the number of honey-producing colonies and yield per colony before 2006, which was when concern over Colony Collapse Disorder spread nationwide?<br>
<b>Q4)</b>Are there any patterns that can be observed between total honey production and the value of production every year?<br>
<b>Q5)</b>How has the value of production, which in some sense could be tied to demand, changed every year?<br>
<b>Q6)</b> Construct the related plots using Seaborn and Matplot apply customization and derive insights from the <br>visualization.

####  Honey Production in USA Dataset:
● <b>state:</b> Various states of the U.S.<br>
●<b> numcol:</b> Number of honey-producing colonies.Honey-producing colonies are the maximum number of colonies from which honey was taken during the year. It is possible to take honey from colonies that did not survive the entire year<br>
● <b>yieldpercol:</b> Honey yield per colony. Unit is pounds<br>
●<b> totalprod:</b> Total production (numcol x yieldpercol). Unit is pounds<br>
●<b> stocks:</b> Refers to stocks held by producers. Unit is pounds<br>
● <b>priceperlb:</b> Refers to an average price per pound based on expanded sales. The unit is dollars.<br>
●<b> prodvalue:</b> Value of production (totalprod x priceperlb). The unit is dollars.<br>
● <b>year:</b> Year of production<br>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Database:

```python
df = pd.read_csv('honeyproduction 1998-2021.csv')
df.head()
```

### Q1) How has honey production yield changed from 1998 to 2021 ?

```python
# Select the relevant columns
d2 = df.groupby('year')['yieldpercol'].mean()
d2.head(25)
```

```python
# Define Figuresize
plt.figure(figsize=(10, 6))

# Create the line plot
sns.lineplot(data=d2, marker='o', linestyle='--', color='m', label='Honey Production Yield')

# Customize the plot
plt.xlabel('Year', color='green', fontsize=12)
plt.ylabel('Production Yield (Million Pounds)', color='green', fontsize=12)
plt.title('Honey Production Yield (1998-2021)', color='darkgreen', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)

# Add some enhancements
sns.set_style("whitegrid")  # Add a white grid background
sns.despine()  # Remove the top and right spines

# Customize the legend
plt.legend(loc='upper left', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

### Insights : 
1. This chart represent the Honey production yield changed.<br>
2. The line plot depicting Honey Production Yield from 1998 to 2021 reveals a consistent upward trend, indicating increased efficiency and expansion in the honey production industry. Notably, there is a temporary dip in production around 2019, likely influenced by external factors, but production subsequently recovers. Seasonal fluctuations may also play a role in yearly variations. Overall, the chart illustrates the industry's resilience and adaptability over the years.
3. from <b>1998 - 2021</b> where the yieldpercol was <b>70</b> in Year <b>1998</b> and the yieldpercol was <b>45</b> in Year <b>2021</b>

### Q2) Overtime, what are the major production trends across the states?

```python
d3 = df.groupby('year')['prodvalue'].sum().reset_index()
d3.head(25)
```

```python
yearly_data = df[df['year'] == 2020]
```

```python
# Sort data by production value in descending order to highlight the major producers
yearly_data = yearly_data.sort_values(by='prodvalue', ascending=False)

# Create a color palette for the bars
colors = plt.cm.Paired(range(len(yearly_data)))

plt.figure(figsize=(12, 6))
bars = plt.bar(yearly_data['State'], yearly_data['prodvalue'],color=colors)
plt.xlabel('State', fontsize=12, color='green')
plt.ylabel('Production Yield (Million Pounds)', fontsize=12, color='green')
plt.title('Honey Production by State in 2020', fontsize=14, color='darkgreen')
plt.xticks(rotation=90, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Add a horizontal line at the average production yield
average_yield = yearly_data['prodvalue'].mean()
plt.axhline(y=average_yield, color='red', linestyle='--', label='Average Yield')

# Highlight the top producer
top_producer = yearly_data.iloc[0]['State']
plt.annotate(f'Top Producer: {top_producer}', xy=(0, 1), xycoords='axes fraction',
             xytext=(10, -20), textcoords='offset points',
             fontsize=10, color='blue', bbox=dict(facecolor='white', alpha=0.8))

plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

### Q3) Does the data show any trends in terms of the number of honey-producing colonies and yield per colony before 2006 , which was when concern over Colony Collapse Disorder spread nation wide?

```python
# Filter data for years before 2006
data_before_2006 = df[df['year'] < 2006]
d4 = data_before_2006.groupby('year')['numcol'].mean().reset_index()
d5 = data_before_2006.groupby('year')['yieldpercol'].mean().reset_index()
```

```python
# Set a customized style for the plots
sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))

# Line chart for the number of colonies
plt.subplot(2, 1, 1)
plt.plot(d4['year'], d4['numcol'], marker='o', linestyle='-', color='b', label='Number of Colonies')
plt.title('Number of Honey-Producing Colonies (Before 2006)',color = 'darkgreen', fontsize=14)
plt.xlabel('Year', color= 'green',fontsize=12)
plt.ylabel('Number of Colonies',color= 'green', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=10, loc='upper left')

# Line chart for yield per colony
plt.subplot(2, 1, 2)
plt.plot(d5['year'], d5['yieldpercol'], marker='o', linestyle='-', color='c', label='Yield Per Colony')
plt.title('Yield Per Colony (Before 2006)',color = 'darkgreen', fontsize=14)
plt.xlabel('Year',color= 'green', fontsize=12)
plt.ylabel('Yield Per Colony (Pounds)',color= 'green', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=10, loc='upper left')
plt.tight_layout()

plt.show()

```

#### Q4) Are there any patterns that can be observed between total honey production and the value of production every year ?

```python
df.head(10)
```

```python
d6 = df.groupby('year')['totalprod'].mean().reset_index()
d7 = df.groupby('year')['prodvalue'].mean().reset_index()
```

```python
plt.figure(figsize=(12, 6))

# Scatter plot of Total Production vs. Value of Production
plt.scatter(df['totalprod'], df['prodvalue'], 
            c=df['year'], cmap='viridis', alpha=0.7, label='Year')

plt.title('Total Production vs. Value of Production (1998-2021)',color = 'darkgreen')
plt.xlabel('Total Production (Million Pounds)', color = 'green')
plt.ylabel('Value of Production (Million Dollars)',color = 'green')
plt.grid(True)

# Add colorbar to show the year
colorbar = plt.colorbar()
colorbar.set_label('Year')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))

# Create a Line plot.
sns.lineplot(data=d6, x='year', y='totalprod', color='blue', marker='o', markersize=8, label='Average Total Production')

plt.title('Average Total Honey Production (1998-2021)', fontsize=16, color='darkgreen')
plt.xlabel('Year', fontsize=12, color='green')
plt.ylabel('Average Total Production (Million Pounds)', fontsize=12, color='green')
plt.xticks(rotation=90)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12, loc='upper left')
plt.tight_layout()

plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

# Create line plot
sns.lineplot(data=d7, x='year', y='prodvalue', marker='d', markersize=8, color='magenta', label='Average Value of Production')
plt.title('Average Value of Honey Production (1998-2021)', fontsize=16, color='darkgreen')
plt.xlabel('Year', fontsize=12, color='green')
plt.ylabel('Average Value of Production (Million Dollars)', fontsize=12, color='green')
plt.xticks(rotation=90)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12, loc='upper left')
plt.tight_layout()

plt.show()
```

### Q5)How has the value of production, which in some sense could be tied to demand, changed every year?

```python
d8= df.groupby('year')['prodvalue'].mean().reset_index()
```

```python
# Set a customized style for the plot
sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))

# Bar chart with enhanced aesthetics
sns.barplot(data=d8, x='year', y='prodvalue', palette='inferno')
plt.title('Average Value of Honey Production by Year (1998-2021)', fontsize=16, color='darkgreen')
plt.xlabel('Year', fontsize=12, color='green')
plt.ylabel('Average Value of Production (Million Dollars)', fontsize=12, color='green')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
